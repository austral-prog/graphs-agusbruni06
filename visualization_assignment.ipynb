# Graficos usando numpy

Primeros pasos con Numpy, dentro de estos plots tenemos pruebas de uso de randoms, arreglos y manejo de dimensiones
Graficos:

    1. Un simple plot combinando "cualquier" valor para x, y
    2. El scatter muestra unicamente los puntos (en enteros) de todas las posibles combinaciones dentro del rango del grafico
    3. Usando la herramienta random de numpy, logramos hacer una linea que no es recta y que sus valores no tienen algun sentido comun o formula
    4. Un histograma representando valores al azar.

Todo esto junto a la combinacion de subplots para poder tener un mejor entendimiento de la libreria

```python
import matplotlib.pyplot as plt
import numpy as np

fig1, axs = plt.subplots(2,2)
x = np.linspace(0, 10, 100)
y = np.sin(x)

axs[0,0].plot(x, y)  # Top-left
axs[0,1].scatter(x, y)  # Top-right
axs[1,0].plot(np.random.random(100).cumsum())
axs[1,1].hist([2,5,6], [1,2,4, 7])

plt.savefig("imgs/graficos.png")
```

# Renderizado en *tres dimensiones*
Usamos matplot para demostrar el grafico con 3 ejes y numpy para el manejo de datos
* Las primeras 3 lineas de codigo (evitando los imports) indican:
1. fig esta creando un grafico de 8x3
2. ax1 esta creando una grilla de 1 fila y 2 columnas
3. ax2 tambien crea una grilla pero mientras que ax1 ocupa la posicion 1, ax2 ocupa la posicion 2

* Valores y variables para las coordenadas
Las primeras 4 variables generadas (_x, _y, _xx y _yy) estan unicamente disenadas para crear una generacion de datos
_xx y _yy estan usando la informacion de _x y de _y para crear una grilla de 4x5 (indicado en el arange de cada variable).
El metodo ravel se usa para aplanar la grilla a una ID obteniendo las coordenadas de que cada linea

* Indicacion del punto mas alto, bajo y tamano
top es una simple suma de x+y que indican cual seria el limite maximo mientras que bottom es el limite minimo
El width simplemente indica la distancia entre cada punto

* Creacion y terminacion del grafico
Ahora pasamos a la parte visible, en donde junto a toda esta informacion dada mostramos dos graficos similares
1. El primer grafico tiene sombreado, esto nos facilita entender que valores tienen cada cosa y nos hace viable la idea de poder verlo en 3 dimensiones
2. El segundo grafico no tiene sombreado, es exactamente igual al primero solo que con esta diferencia de no poder distinguir las dimensiones

Si quieren saber como aprendi estas cosas pueden fijarse en su [demo](https://matplotlib.org/stable/gallery/mplot3d/3d_bars.html) o directamente aprender desde la [documentacion](https://matplotlib.org/stable/gallery/mplot3d/index.html)

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(8,3))
ax1 = fig.add_subplot(121, projection="3d")
ax2 = fig.add_subplot(122, projection="3d")

_x = np.arange(4)
_y = np.arange(5)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()

top = x + y
bottom = np.zeros_like(top)
width = depth = 1

ax1.bar3d(x, y, bottom, width, depth, top, shade=True)
ax1.set_title("Sombreado")

ax2.bar3d(x, y, bottom, width, depth, top, shade=False)
ax2.set_title("Sin sombrear")

plt.savefig("imgs/tres_dimensiones.png")
plt.show()
```

# Funcion Trigonometrica
Usamos funciones matematicas gracias a numpy como el seno, coseno para indicar como tendria que moverse el grafico.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 300)

plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), label="sin(x)", color="steelblue", linewidth=2)
plt.plot(x, np.cos(x), label="cos(x)", color="coral", linewidth=2, linestyle="--")

plt.axhline(0, color="gray", linewidth=0.8, linestyle=":")
plt.title("Funciones trigonométricas")

plt.xlabel("x (radianes)")
plt.ylabel("y")

plt.savefig("imgs/func_trigonometricas.png")
plt.legend()
plt.tight_layout()
plt.show()
```

# Asintotas verticales y horizontales

En este caso, seguimos usando funciones aparte para crear asintotas, el metodo de concatenar nos permite trabajar muy cerquita del cero, de tal forma que nunca llegue a el y nos produzcan errores


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.concatenate([np.linspace(-5, -0.1, 300), np.linspace(0.1, 5, 300)])
y = 1 / x

plt.figure(figsize=(8, 5))
plt.plot(x, y, color="steelblue", linewidth=2, label="f(x) = 1/x")
plt.axvline(0, color="coral", linewidth=1.2, linestyle="--", label="asíntota vertical x=0")
plt.axhline(0, color="gray", linewidth=1.2, linestyle="--", label="asíntota horizontal y=0")
plt.ylim(-10, 10)
plt.title("f(x) = 1/x — asíntotas vertical y horizontal")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.tight_layout()

plt.savefig("imgs/asintotas.png")
plt.show()
```
